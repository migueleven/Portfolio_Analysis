<h2 style="color: #FF6347;">Loading Data.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
portfolio_weights = pd.read_csv("portfolio_weights.csv")

portfolio_weights['date'] = pd.to_datetime(portfolio_weights['date'])

# Get years
portfolio_weights['Year'] = portfolio_weights['date'].dt.year

# Change index
portfolio_weights.set_index('date', inplace=True)

portfolio_weights.head()
```

<h2 style="color: #FF6347;">Area Chart of Asset Weights.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
plt.figure(figsize=(10,6))

for col in asset_data.columns:
    plt.stackplot(portfolio_weights.index, portfolio_weights[col], labels=[col])
    plt.legend()
```

<h2 style="color: #FF6347;">Chart of Historical Cumulative Returns of the Portfolio.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
plt.figure(figsize=(10, 6))

# Calcular la suma de cada columna
column_sums = portfolio_weights.drop(['Year'], axis=1).sum()

print(column_sums.head())

# Crear el gráfico de barras
plt.bar(x=column_sums.index, height=column_sums.values)

# Opcional: Estilizar el gráfico
plt.xlabel("Portfolio Components")
plt.ylabel("Weight Sum")
plt.title("Portfolio Weights Distribution")
plt.show()
    
```

<h2 style="color: #FF6347;">Annualized Return.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
# Paso 1: Transformar los datos a formato "long" (melt) para Seaborn
portfolio_melted = portfolio_weights.melt(id_vars='Year', 
                                          var_name='Asset', 
                                          value_name='Weight')

# Paso 2: Crear el gráfico de barras agrupado usando Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Weight', hue='Asset', data=portfolio_melted, palette='muted')

# Paso 3: Ajustar el gráfico
plt.title('Portfolio Weights by Year and Asset')
plt.xlabel('Year')
plt.ylabel('Weight')
plt.legend(title='Assets')
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

<h2 style="color: #FF6347;">Annualized Volatility.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
# Calcular retornos diarios
daily_returns = asset_data.pct_change()

# Calcular volatilidad anualizada
annualized_volatility = daily_returns.std() * np.sqrt(252)

annualized_volatility.plot(kind='bar', title="Annualized Volatility of Assets", figsize=(8, 6))
plt.ylabel("Volatility")
plt.show()
```
